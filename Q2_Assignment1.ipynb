```python
from google.colab.patches import cv2_imshow
from skimage.io import imread,imshow
import cv2
import numpy as np
```

# First - order Techniques (Sobel)


```python
img2=imread('tank.jpg',0)
kernel = np.array([[0,0,0 ],
 [0,0,-1 ],
 [0,1 ,0]])
m,n=img2.shape
out=np.zeros([m,n])
for i in range(1,m-1):
 for j in range(1,n-1):
  temp=0
  for a in range(-1,2):
   for b in range(-1,2):
    temp+=img2[i+a,j+b]*kernel[a+1,b+1]
    out[i,j]=temp
    kernel2 = np.array([[0,0,0 ],
 [0,-1,0 ],
 [0,0 ,1]])
m,n=img2.shape
out1=np.zeros([m,n])
for i in range(1,m-1):
 for j in range(1,n-1):
  temp=0
  for a in range(-1,2):
   for b in range(-1,2):
    temp+=img2[i+a,j+b]*kernel[a+1,b+1]
    out1[i,j]=temp
cv2_imshow(out+out1)
cv2.waitKey()
cv2.destroyAllWindows()
```

#Second Order Techniques
# Unsharpening mask

```python
img2=imread('tank.jpg',0)
kernel = np.array([[1/9,1/9,1/9],
 [1/9,1/9,1/9],
 [1/9,1/9,1/9]])
m,n=img2.shape
out=np.zeros([m,n])
for i in range(1,m-1):
 for j in range(1,n-1):
  temp=0
  for a in range(-1,2):
   for b in range(-1,2):
    temp+=img2[i+a,j+b]*kernel[a+1,1+b]
    out[i,j]=temp
    mean_im=img2 - out
    final=mean_im + img2
cv2.imshow(final)
cv2.waitKey()
cv2.destroyAllWindows()
```
